### Impute missing values

# Handle Categorical Encoding

```python
trainSet.head()
```

```python
trainSet_enc = pd.get_dummies(columns=['v1', 'v4', 'v8', 'v9', 'v11', 'v12'], data=trainSet, drop_first=True)
```

```python
trainSet_enc.head()
```

__Encoding Y: __  In column classLabel encoding the minority class - No - as 1 and majority class - Yes - as 0

```python
trainSet_enc['classLabel'] = pd.to_numeric(trainSet_enc.classLabel.str.replace('yes.', '0').str.replace('no.', '1'))
```

```python
trainSet_enc.head()
```

```python
trainSet_enc.classLabel.value_counts()
```

```python
trainSet[trainSet['v17'] != trainSet['classLabel']].shape
```

```python
sns.pairplot(trainSet)
```

__Observations__
- V17 is the numeric encoded version of classLabel, so will have to removed from the X data
- V2, V3, V7, V10, and V13 are all right skewed, so there *might be* some benefit from a log transformation to make it more normal for certain classifiers

## Split into X and Y

```python
trY = trainSet_enc.classLabel
trX = trainSet_enc.drop('classLabel', axis=1)
```

```python
trX.head()
```

## Base performance
Since the dataset is highly unbalanced, I just wanted to see what the scores would look like if we just returned Yes for every row

```python
y_pred_always_yes = np.zeros(trX.shape[0])
```

```python
tn, fp, fn, tp = confusion_matrix(y_true=trY_enc, y_pred=y_pred_always_yes).ravel()
print("tp, fp, fn, tn:", tp, fp, fn, tn)
print("tp, fp, fn, tn as %s:", np.array([tp, fp, fn, tn])*100/trX.shape[0])
print("F1 Score:", f1_score(y_true=trY_enc, y_pred=y_pred_always_yes))
print("Accuracy:", accuracy_score(y_true=trY_enc, y_pred=y_pred_always_yes))
```
